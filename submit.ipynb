```python
##处理OSV数据，将其中不规则的数据进行标准化处理
def osv_trans(x):
    x = str(x).replace('Android_','').replace('Android ','').replace('W','')
    if str(x).find('.') > 0:
        temp_index1 = x.find('.')
        if x.find(' ') > 0:
            temp_index2 = x.find('')
        else:
            temp_index2 = len(x)
        if x.find('-') > 0:
            temp_index2 = x.find('-')
        result = x[0:temp_index1] +'.'+x[temp_index1 + 1:temp_index2].replace('.','')
        try:
            return float(result)
        except:
            #print(x+"###")
            return 0
    try:
        return float(x)
    except:
        #print(x + "###")
        return 0
train['osv'] = train['osv'].apply(osv_trans)
test['osv'] = test['osv'].apply(osv_trans)

    
```

```python
#构建特征比，生成两种标签数据中OSV的特征比的数据信息
def find_key_feature(train,selected):
    temp0 = train[train['label'] == 0]
    temp1 = train[train['label'] == 1]
    temp = pd.DataFrame(columns=[0,1])
    temp[0] = temp0['osv'].value_counts() / len(temp0) * 100
    temp[1] = temp1['osv'].value_counts() / len(temp1) * 100
    temp[2] = temp[1] / temp[0]
    #筛选大于10的关键特征
    result = temp[temp[2] > 10].sort_values(2,ascending=False).index
    return result
key_feature= {}
key_feature['osv'] = find_key_feature(train,'osv')
key_feature
```

```python
#5折交叉验证，结合lightGBM以及XGboost，参考自：https://aistudio.baidu.com/paddle/forum/topic/show/994388
from lightgbm import LGBMClassifier
import xgboost as xgbc
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score,roc_auc_score

def lgb_xgb_train(KF,X,y,X_test):
    seed = 2022
    lgb = LGBMClassifier(num_leaves=1024,
                           max_depth=12,
                           learning_rate=0.005,
                           n_estimators=5000,
                           subsample=0.8,
                           feature_fraction=0.8,
                           reg_alpha=0.5,
                           reg_lambda=0.5,
                           random_state=seed,
                           metric='auc',
                           boosting_type='gbdt',
                           subsample_freq=1,
                           device='gpu', 
                           bagging_fraction=0.8,verbose=-1)
    
    xgb=xgbc.XGBClassifier(
            max_depth=13, learning_rate=0.005, n_estimators=2400, 
            objective='binary:logistic',tree_method='gpu_hist',
            subsample=0.95, colsample_bytree=0.4,n_jobs=-1,random_state=seed, 
            min_child_samples=3, eval_metric='auc', reg_lambda=0.5,verbosity=0)
    
    xgb_prob = np.zeros(X_test.shape[0])
    lgb_prob = np.zeros(X_test.shape[0])
    skf = StratifiedKFold(n_splits=KF, shuffle=True, random_state=2022)

    for k,(train_index, test_index) in enumerate(skf.split(X, y)):
        X_train, X_val = X.iloc[train_index], X.iloc[test_index]
        y_train, y_val = y.iloc[train_index], y.iloc[test_index]
        print(f'第{k+1}次训练')
        
        # 训练
        xgb = xgb.fit(X_train,y_train,
                      eval_set=[(X_val,y_val)],
                      verbose=False)
        lgb = lgb.fit(X_train,y_train,
                      eval_metric='auc',
                              verbose = False)
        
        # 正式预测
        xgb_prob += xgb.predict_proba(X_test)[:,1]/KF
        print(f'Xgboost:第{k+1}次训练的AUC{roc_auc_score(y_val,xgb.predict_proba(X_val)[:,1])},第{k+1}次训练的Accuracy{accuracy_score(y_val,xgb.predict(X_val))}')
        lgb_prob += lgb.predict_proba(X_test)[:,1]/KF
        print(f'LightGBM:第{k+1}次训练的AUC{roc_auc_score(y_val,lgb.predict_proba(X_val)[:,1])},第{k+1}次训练的Accuracy{accuracy_score(y_val,lgb.predict(X_val))}')
    
    return xgb_prob,lgb_prob
```

```python
#代码预测结果
xgb_prob,lgb_prob = lgb_xgb_train(5,features.drop(['timestamp'],axis=1),train['label'],test_features.drop(['timestamp'],axis = 1))
```

```python
#预测结果预览
result = test[['sid']]
result['xgb_prob'] = xgb_prob
result['lgb_prob'] = lgb_prob
result['proba'] = xgb_prob + lgb_prob
result['label'] = result['proba'].apply(lambda x:1 if x>=1 else 0)

#Tickid中的数据已经是在训练集中有对应标签，即已经得到结果
Trick_id = [1260961, 1393628, 1356887, 1911347,1298847, 1629795]
result[result['sid'].apply(lambda x:x in Trick_id)]['xgb_prob'] = 1
result[result['sid'].apply(lambda x:x in Trick_id)]['lgb_prob'] = 1

result
```

### Summary：

本次比赛为反欺诈风险预测，一共提供50万的训练数据以及15万的测试数据。
1.在数据预处理阶段，首先对数据进行了清洗，因为在OSV，LAN等列中包含Nan以及不标准的数据，对于空值，根据统计结果采用了众数填补的处理方法，而对于不规则的数据，则写了函数来进行处理。同时针对字符数据，例如feahash这种则进行了长度计算使之能够成为建模的特征数据。

3.训练分别采用了LightGBM，XGboost以及神经网络模型，同时采用了模型融合和5折交叉验证的方法，单一模型的得分在0.86-0.89之间，而效果最好的则是使用LightGBM以及XGBoost进行模型融合，并进行五折交叉验证最终得出的结果。
